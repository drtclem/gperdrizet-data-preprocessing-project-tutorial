```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
raw_data = pd.read_csv("https://raw.githubusercontent.com/4GeeksAcademy/data-preprocessing-project-tutorial/main/AB_NYC_2019.csv")

raw_data
```

```python
raw_data.describe()
```

```python
raw_data.info()
```

```python
fig, axis = plt.subplots(figsize=(10, 7), nrows=2)
axis = axis.flatten()

variables = ['neighbourhood_group', 'room_type']

for i, variable in enumerate(variables):
    sns.countplot(ax=axis[i], data=raw_data, x=variable, hue=variable)


plt.tight_layout()

plt.show()
```

```python
fig, axis = plt.subplots(figsize=(10, 7), nrows=2)
axis = axis.flatten()

variables = ['room_type', 'reviews_per_month']

for i, variable in enumerate(variables):
    sns.countplot(ax=axis[i], data=raw_data, x='neighbourhood_group', hue=variable)


plt.tight_layout()

plt.show()
```

```python

variables = ['price', 'minimum_nights', 'availability_365']

for i, variable in enumerate(variables):
    sns.displot(data=raw_data, x=variable, hue='neighbourhood_group', kind='kde')
```

```python
raw_data.columns
```

```python

variables = ['number_of_reviews', 'minimum_nights', 'availability_365']
for variable in variables:
    sns.relplot(data=raw_data, x='price', y=variable)
```

```python
variables = ['room_type', 'neighbourhood_group']

for variable in variables:
    sns.catplot(data=raw_data, x=variable, y='price', hue='neighbourhood_group')
    sns.catplot(data=raw_data, x=variable, y='price', hue='neighbourhood_group', kind='box', log_scale=True)
```
